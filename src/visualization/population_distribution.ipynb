```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import openpyxl
```

```python
filepath = '../../data/raw/df.xlsx'
df = pd.read_excel(filepath)

df['population'] = pd.to_numeric(df['population'], errors='coerce')

populations = df['population'].dropna()
```

```python
populations.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(populations, kde=True)
plt.title('Population Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(np.log(df['population']+1))
plt.title('Log Population Distribution + 1')
plt.show()
```

```python
population_change_rate = df.groupby('island')['population'].pct_change()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['year'], population_change_rate)
plt.show()
```
