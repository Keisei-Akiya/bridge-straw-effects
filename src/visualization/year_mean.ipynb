# 年平均の推移

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
import seaborn as sns

figsize = (10, 5)
```

```python
path = '../../data/processed/df_filtered.xlsx'
df = pd.read_excel(path)
```

```python
year_mean = df.groupby('year')['log_pop'].mean()
year_mean_diff = year_mean.diff()
year_mean_diff.fillna(0, inplace=True)
year_mean_cumsum = year_mean_diff.cumsum()
year_mean_change_rate = 100 * (np.exp(year_mean_cumsum) - 1)
```

```python
fig, ax = plt.subplots(figsize=figsize)
sns.lineplot(data=year_mean_change_rate, ax=ax, color='black', marker='o')
ax.set_title('年平均の変化率推移')
ax.set_ylabel('変化率 (%)')
ax.set_xlabel('年')
```
