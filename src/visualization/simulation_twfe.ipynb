```python
import pandas as pd
import numpy as np
import japanize_matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# サンプルデータのサイズ
n_pre = 500  # 介入前の期間
n_post = 500  # 介入後の期間

# 介入前のデータを生成
time_pre = np.linspace(0, n_pre-1, n_pre)
control_pre = np.linspace(100, 50, n_pre)  # コントロールグループの単調減少
treatment_pre = np.linspace(110, 60, n_pre)  # 介入グループの単調減少

# 介入後のデータを生成
time_post = np.linspace(n_pre, n_pre + n_post - 1, n_post)
control_post = np.linspace(50, 0, n_post)  # コントロールグループの単調減少
treatment_post = np.linspace(60, 30, n_post)  # 介入グループの単調減少（介入効果で減少が緩やか）
treatment_post_if_control = np.linspace(60, 20, n_post)  # 介入グループのが介入を受けなかった場合

# データを結合
time = np.concatenate([time_pre, time_post])
control = np.concatenate([control_pre, control_post])
treatment = np.concatenate([treatment_pre, treatment_post])
treatment_post_if_control = np.concatenate([treatment_pre, treatment_post_if_control])

# データフレームを作成
data = pd.DataFrame({
    'time': time,
    'control': control,
    'treatment': treatment,
    'treatment_post_if_control': treatment_post_if_control
})

# データをプロット
plt.figure(figsize=(12, 6))
plt.plot(data['time'], data['control'], label='対照群', color='black')
plt.plot(data['time'], data['treatment'], label='介入郡', color='black')
plt.plot(data['time'], data['treatment_post_if_control'], label='介入郡が介入を受けなかった場合', color='black', linestyle='--')
plt.axvline(x=n_pre-1, color='black', linestyle=':', label='介入開始')
plt.title('橋が人口減少を抑えるケース')
plt.xlabel('時間')
plt.ylabel('人口')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# サンプルデータのサイズ
n_pre = 500  # 介入前の期間
n_post = 500  # 介入後の期間

# 介入前のデータを生成
time_pre = np.linspace(0, n_pre-1, n_pre)
control_pre = np.linspace(80, 30, n_pre)  # コントロールグループの単調減少
treatment_pre = np.linspace(110, 60, n_pre)  # 介入グループの単調減少

# 介入後のデータを生成
time_post = np.linspace(n_pre, n_pre + n_post - 1, n_post)
control_post = np.linspace(30, -20, n_post)  # コントロールグループの単調減少

# 介入グループの段階的な効果を追加
treatment_post = np.linspace(60, 30, n_post)  # 基本的な単調減少
treatment_post_if_control = np.linspace(60, 20, n_post)  # 介入グループが介入を受けなかった場合

# 段階的な効果を追加
effect_size_negative = -10  # マイナスの効果の大きさ
effect_size_positive = 10  # プラスの効果の大きさ
effect_steps_negative = 2  # マイナスの効果が現れるステップ数
effect_steps_positive = 3  # プラスの効果が現れるステップ数

# マイナスの効果を追加
for i in range(effect_steps_negative):
    start = i * (n_post // (effect_steps_negative + effect_steps_positive))
    end = (i + 1) * (n_post // (effect_steps_negative + effect_steps_positive))
    treatment_post[start:end] += effect_size_negative * (i + 1) / effect_steps_negative

# プラスの効果を追加
for i in range(effect_steps_positive):
    start = effect_steps_negative * (n_post // (effect_steps_negative + effect_steps_positive)) + i * (n_post // (effect_steps_negative + effect_steps_positive))
    end = effect_steps_negative * (n_post // (effect_steps_negative + effect_steps_positive)) + (i + 1) * (n_post // (effect_steps_negative + effect_steps_positive))
    treatment_post[start:end] += effect_size_positive * (i + 1) / effect_steps_positive

# もっと前に介入を受けたグループのデータを生成
early_treatment_pre = np.linspace(140, 100, n_pre-100)  # 介入前のデータ
early_treatment_post = np.linspace(100, 80, n_post+100)  # 介入後のデータ（介入効果で減少が緩やか）
early_treatment_post_if_control = np.linspace(100, 50, n_post+100)  # 介入を受けなかった場合

# マイナスの効果を追加
for i in range(effect_steps_negative):
    start = i * (n_post // (effect_steps_negative + effect_steps_positive))
    end = (i + 1) * (n_post // (effect_steps_negative + effect_steps_positive))
    early_treatment_post[start:end] += effect_size_negative * (i + 1) / effect_steps_negative

# プラスの効果を追加
for i in range(effect_steps_positive):
    start = effect_steps_negative * (n_post // (effect_steps_negative + effect_steps_positive)) + i * (n_post // (effect_steps_negative + effect_steps_positive))
    end = effect_steps_negative * (n_post // (effect_steps_negative + effect_steps_positive)) + (i + 1) * (n_post // (effect_steps_negative + effect_steps_positive))
    early_treatment_post[start:end] += effect_size_positive * (i + 1) / effect_steps_positive

# データを結合
time = np.concatenate([time_pre, time_post])
control = np.concatenate([control_pre, control_post])
treatment = np.concatenate([treatment_pre, treatment_post])
treatment_post_if_control = np.concatenate([treatment_pre, treatment_post_if_control])
early_treatment = np.concatenate([early_treatment_pre, early_treatment_post])
early_treatment_post_if_control = np.concatenate([early_treatment_pre, early_treatment_post_if_control])

# データフレームを作成
data = pd.DataFrame({
    'time': time,
    'control': control,
    'treatment': treatment,
    'treatment_post_if_control': treatment_post_if_control,
    'early_treatment': early_treatment,
    'early_treatment_post_if_control': early_treatment_post_if_control
})

# データをプロット
plt.figure(figsize=(12, 6))
plt.plot(data['time'], data['early_treatment'], label='第1介入郡', color='red')
plt.plot(data['time'], data['early_treatment_post_if_control'], label='第1介入郡が介入を受けなかった場合', color='red', linestyle='--')
plt.plot(data['time'], data['treatment'], label='第2介入郡', color='blue')
plt.plot(data['time'], data['treatment_post_if_control'], label='第2介入郡が介入を受けなかった場合', color='blue', linestyle='--')
plt.plot(data['time'], data['control'], label='対照群', color='black')
plt.axvline(x=n_pre-100, color='red', linestyle=':')
plt.axvline(x=n_pre-1, color='blue', linestyle=':')
plt.title('介入効果が時間変動するケース')
plt.xlabel('時間')
plt.ylabel('人口')
plt.legend()
plt.show()
```
