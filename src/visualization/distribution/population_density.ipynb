# 人口密度の分布

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
filepath = '../../../data/processed/df_filtered_5years.xlsx'
df = pd.read_excel(filepath)
df['density'] = df['population'] / df['area_km2']
```

```python
df['density'].describe()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['density'], kde=True);
```

```python
log_density = np.log(df['density'] + 1)

plt.figure(figsize=(12, 6))
sns.histplot(log_density, kde=True);
```

```python
standardized_density = (log_density - log_density.mean()) / log_density.std()

plt.figure(figsize=(12, 6))
sns.histplot(standardized_density, kde=True);
```
