```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

filepath = '../../../data/processed/df_filtered_5years.xlsx'
df = pd.read_excel(filepath)
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['income'], kde=True)
plt.title('Income Distribution')
plt.show()
```

## 対数

```python
df['log_income'] = np.log(df['income'])
plt.figure(figsize=(10, 6))
sns.histplot(df['log_income'], kde=True)
plt.title('Log Income Distribution')
plt.show()
```

## 標準化

```python
standardized_log_income = (df['log_income'] - df['log_income'].mean()) / df['log_income'].std()
plt.figure(figsize=(10, 6))
sns.histplot(standardized_log_income, kde=True)
plt.title('Standardized Log Income Distribution');
```

```python
log_income_diff = df.groupby('island_id')['log_income'].diff().dropna()
plt.figure(figsize=(12, 6))
sns.histplot(log_income_diff, kde=True);
```
