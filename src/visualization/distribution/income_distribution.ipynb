```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filepath = '../../data/processed/df_filtered_5years.xlsx'
df = pd.read_excel(filepath)
income = df['income']
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(income, kde=True)
plt.title('Income Distribution')
plt.show()
```

```python
log_income = np.log(income)
plt.figure(figsize=(10, 6))
sns.histplot(log_income, kde=True)
plt.title('Log Income Distribution')
plt.show()
```

```python
standardized_income = (income - income.mean()) / income.std()
plt.figure(figsize=(10, 6))
sns.histplot(standardized_income, kde=True)
plt.title('Standardized Income Distribution');
```

```python
standardized_log_income = (log_income - log_income.mean()) / log_income.std()
plt.figure(figsize=(10, 6))
sns.histplot(standardized_log_income, kde=True)
plt.title('Standardized Log Income Distribution');
```
