# 人口 分布

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import openpyxl
```

```python
filepath = '../../data/processed/df_filtered_5years.xlsx'
df = pd.read_excel(filepath)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['population'], kde=True)
plt.title('Population Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(np.log(df['population']+1))
plt.title('Log Population Distribution + 1')
plt.show()
```

```python
population_change_rate = df.groupby('island')['population'].pct_change()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['year'], population_change_rate)
plt.show()
```
