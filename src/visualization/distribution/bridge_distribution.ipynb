```python
import pandas as pd
import numpy as np
import japanize_matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filepath = '../../../data/processed/df_filtered.xlsx'
df = pd.read_excel(filepath)
```

```python
bridge_year_list = df['bridge_year'].unique()
bridge_year_list.sort()
bridge_year_list
```

```python
paseudo_bridge_year_list = df['paseudo_bridge_year'].unique()
paseudo_bridge_year_list.sort()
paseudo_bridge_year_list
```

```python
for bridge_year in bridge_year_list:
    num_of_bridge = len(df[(df['year'] == 2020) & (df['bridge_year'] == bridge_year)])
    print(f'{bridge_year}年度: {num_of_bridge}件')
```

```python
for paseudo_bridge_year in paseudo_bridge_year_list:
    num_of_bridge = len(df[(df['year'] == 2020) & (df['paseudo_bridge_year'] == paseudo_bridge_year)])
    print(f'{paseudo_bridge_year}年度: {num_of_bridge}件')
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df[df['year'] == 2020]['bridge_year'], bins=400, color='black')
plt.title('離島架橋 建設年別 本数')
plt.ylim(0, 8)
plt.xlim(1930, 2020)
plt.ylabel('本数')
plt.xlabel('年')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df[df['paseudo_year'] == 2020]['paseudo_bridge_year'], bins=400, color='black')
plt.title('離島架橋 建設年別 本数')
plt.ylim(0, 8)
plt.xlim(1930, 2020)
plt.ylabel('本数')
plt.xlabel('年')
plt.show()
```
