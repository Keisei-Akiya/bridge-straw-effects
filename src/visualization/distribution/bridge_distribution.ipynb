```python
import pandas as pd
import numpy as np
import japanize_matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams
```

```python
filepath = '../../../data/processed/df_filtered.xlsx'
df = pd.read_excel(filepath)
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df[df['years'] == 2020]['year_bridges_opened'], bins=400, color='black')
plt.title('離島架橋 建設年別 本数')
plt.ylim(0, 8)
plt.xlim(1930, 2020)
plt.ylabel('本数')
plt.xlabel('年')
plt.show()
```
