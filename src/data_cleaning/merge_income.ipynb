# 人口データと課税所得データを結合する．

```python
import pandas as pd
import numpy as np
import openpyxl
```

```python
df = pd.read_excel('../../data/raw/df.xlsx')
df = df.where(pd.notnull(df), None)
df
```

```python
df_income = pd.read_csv('../../data/raw/taxable_income.csv', encoding='utf-8')

df['region_code'] = df['region_code'].astype(float)
df_income['region_code'] = df_income['region_code'].astype(float)
```

```python
merged_df = pd.merge(df, df_income[['year', 'region_code', 'taxable_income']], on=['year', 'region_code'], how='left')
# Step 1: Check for missing values in 'income' and 'taxable_income' columns
# Step 2: Fill missing values in 'income' with values from 'taxable_income'
# Step 3: Fill missing values in 'taxable_income' with values from 'income'
# Step 4: Create a new column 'combined_income' with the filled values

merged_df['combined_income'] = merged_df['income'].combine_first(merged_df['taxable_income'])
merged_df
```

```python
df['income'] = merged_df['combined_income']
df
```

```python
output_filepath = '../../data/raw/df.xlsx'
df.to_excel(output_filepath, index=False)
```
