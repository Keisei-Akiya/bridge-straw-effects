# 過去のデータを入れるための列を追加する

## 方針

1. 1995年のデータをコピーする．
2. year列を2000, 1995に書き換える．
3. population, income列を空白にする．
4. 結合する．

```python
import pandas as pd
import numpy as np
import openpyxl
```

```python
filepath = '../../data/raw/df.xlsx'
df = pd.read_excel(filepath, engine='openpyxl')
df
```

```python
df_1995 = df[df['year'] == 1995]
df_1995.head()
```

```python
df_1990 = df_1995.copy()
df_1990['year'] = 1990
df_1990['population'] = ''
df_1990['income'] = ''
df_1990 = df_1990[~df_1990.set_index(['island_id', 'year']).index.isin(df.set_index(['island_id', 'year']).index)]
df_1990.head()
```

```python
df_1985 = df_1995.copy()
df_1985['year'] = 1985
df_1985['population'] = ''
df_1985['income'] = ''
df_1985 = df_1985[~df_1985.set_index(['island_id', 'year']).index.isin(df.set_index(['island_id', 'year']).index)]
df_1985.head()
```

```python
df_1980 = df_1995.copy()
df_1980['year'] = 1980
df_1980['population'] = ''
df_1980['income'] = ''
df_1980 = df_1980[~df_1980.set_index(['island_id', 'year']).index.isin(df.set_index(['island_id', 'year']).index)]
df_1980.head()
```

```python
merged_df = pd.concat([df, df_1980, df_1985, df_1990, df_1995], ignore_index=True)
merged_df = merged_df.sort_values(['island_id', 'year']).reset_index(drop=True)
merged_df
```

```python
output_filepath = '../../data/raw/merged_df.xlsx'
merged_df.to_excel(output_filepath, index=False)
```
