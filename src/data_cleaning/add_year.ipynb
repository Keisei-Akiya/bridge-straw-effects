# 過去のデータを入れるための列を追加する

## 方針

1. 2001年のデータをコピーする．
2. year列を2000, 1995に書き換える．
3. population, income列を空白にする．
4. 結合する．

```python
import pandas as pd
import numpy as np
import openpyxl
```

```python
filepath = '../../data/raw/df.xlsx'
df = pd.read_excel(filepath, engine='openpyxl')
df.head()
```

```python
df_2001 = df[df['year'] == 2001]
df_2001.head()
```

```python
df_2000 = df_2001.copy()
df_2000['year'] = 2000
df_2000['population'] = ''
df_2000['income'] = ''
df_2000 = df_2000[~df_2000.set_index(['island_id', 'year']).index.isin(df.set_index(['island_id', 'year']).index)]
df_2000.head()
```

```python
df_1995 = df_2001.copy()
df_1995['year'] = 1995
df_1995['population'] = ''
df_1995['income'] = ''
df_1995 = df_1995[~df_1995.set_index(['island_id', 'year']).index.isin(df.set_index(['island_id', 'year']).index)]
df_1995.head()
```

```python
df = pd.concat([df, df_2000, df_1995], ignore_index=True)
df = df.sort_values(['island_id', 'year']).reset_index(drop=True)
df
```

```python
output_filepath = '../../data/raw/df_add_1995_2000.xlsx'
df.to_excel(output_filepath, index=False)
```
