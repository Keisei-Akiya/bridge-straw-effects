# 過去のデータを入れるための列を追加する

## 方針

1. 1995年のデータをコピーする．
2. year列を2000, 1995に書き換える．
3. population, income列を空白にする．
4. 結合する．

```python
import pandas as pd
import numpy as np
```

```python
filepath = '../../data/raw/df.xlsx'
df = pd.read_excel(filepath)
df
```

```python
df_1975 = df[df['year'] == 1975]
df_1975.head()
```

```python
df_1970 = df_1975.copy()
df_1970['year'] = 1970
df_1970['population'] = ''
df_1970['income'] = ''
df_1970 = df_1970[~df_1970.set_index(['island_id', 'year']).index.isin(df.set_index(['island_id', 'year']).index)]
df_1970.head()
```

```python
merged_df = pd.concat([df, df_1970], ignore_index=True)
merged_df = merged_df.sort_values(['island_id', 'year']).reset_index(drop=True)
merged_df
```

```python
output_filepath = '../../data/raw/df.xlsx'
merged_df.to_excel(output_filepath, index=False)
```
