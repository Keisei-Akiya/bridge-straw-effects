# 島一覧取得

```python
import pandas as pd
import numpy as np

filepath = '../../data/raw/df.xlsx'
df = pd.read_excel(filepath)
```

```python
df.head()
```

```python
islands_list = df[['island', 'island_id', 'region_code', 'prefecture_code']].drop_duplicates().reset_index(drop=True)
islands_list
```

```python
output_filepath = '../../data/raw/islands_list.xlsx'
islands_list.to_excel(output_filepath, index=False)
```
