# 面積

面積を島リストに転記

```python
import pandas as pd

islands_list_filepath = '../../data/raw/islands_list.xlsx'
islands_area_filepath = '../../data/raw/islands_area.xlsx'
prefecture_code_filepath = '../../data/processed/prefecuture_code.csv'
df_island_list = pd.read_excel(islands_list_filepath)
df_island_area = pd.read_excel(islands_area_filepath)
df_prefecture_code = pd.read_csv(prefecture_code_filepath)
```

```python
df_island_list.info()
```

```python
df_island_area[df_island_area['island_name'] == '大島']
```

```python
df_prefecture_code['prefecture_name'] = df_prefecture_code[' prefecture_name'].str.replace(' ', '')
df_prefecture_code
```

```python
df_merged_area_code = df_island_area.merge(df_prefecture_code[['prefecture_code', 'prefecture_name']], on='prefecture_name', how='left')
df_merged_area_code
```

```python
# Merge the dataframes on the corresponding columns
df_merged = df_island_list.merge(
    df_merged_area_code[['island_name', 'area', 'prefecture_code']],
    left_on=['island', 'prefecture_code'],
    right_on=['island_name', 'prefecture_code'],
    how='left'
)

# Drop the redundant 'island_name' column
df_merged.drop(columns=['island_name'], inplace=True)
```

```python
# Display the merged dataframe
df_merged.head()
output_filepath = '../../data/raw//merged_island_list.xlsx'
df_merged.to_excel(output_filepath, index=False)
```
