# 遠すぎる島を除外する．

世界で最も長い吊橋としてギネス記録に認定されているのは．明石海峡大橋．

この橋は中央支間が1,991メートル．

これより距離がある島に橋をかけるのは技術的に困難であるため，遠すぎる島をサンプルから排除する．

```python
import pandas as pd
import numpy as np

filepath = '../../data/raw/df.xlsx'
df = pd.read_excel(filepath)
df.head()
```

```python
df_exclude_too_far_islands = df[df['distance_m'] <= 1991]
```

```python
output_filepath = '../../data/raw/df_exclude_too_far_islands.xlsx'
df_exclude_too_far_islands.to_excel(output_filepath, index=False)
```
