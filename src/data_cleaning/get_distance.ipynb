```python
import pandas as pd

df_filepath = '../../data/raw/df.xlsx'
islands_list_filepath = '../../data/raw/islands_list.xlsx'
df = pd.read_excel(df_filepath)
islands_list = pd.read_excel(islands_list_filepath)
```

```python
df['prefecture_name'] = df['prefecture_code.1']
df = df.drop(columns=['prefecture_code.1'])
df.head()
```

```python
islands_list.head()
```

```python
merged_df = pd.merge(df, islands_list[['island_id', 'area_km2', 'distance_m']], on='island_id')
merged_df.head()
```

```python
export_filepath = '../../data/raw/df.xlsx'
merged_df.to_excel(export_filepath, index=False)
```
